```python
#붓꽃 데이터 읽어와서 확인하기
import pandas as pd
df=pd.read_csv('iris.csv')
#기초 정보
df.info()
#기초 통계량
df.describe()
#품종별 개수 구하기
species_count = df.species.value_counts()
print(species_count)
```

```python
#붓꽃 데이터 그룹핑
species_sum = df.groupby('species').sum()
print(species_sum)
species_mean = df.groupby('species').mean()
print(species_mean)
```

```python
#붓꽃 데이터 시각화
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
#꽃받침 길이(Sepal length)의 시각화(이거 수정 필요)
plt.figure()
sns.barplot(x='species', y='sepal_length', data=df)
plt.title('Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
#붓꽃 종류별 꽃받침 길이의 평균에 대한 시각화
sepal_length_mean = df.groupby('species').mean()
sns.barplot(x='species', y='sepal_length', data=sepal_length_mean)
plt.title('Mean Sepal Length')
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length')
plt.show()
```

```python
#히스토그램 그래프 그리기
#각 특징(4가지) 값들의 빈도수에 대한 시각화
import seaborn as sns
import matplotlib.pyplot as plt

# 특징들의 빈도수에 대한 시각화
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
sns.histplot(df['sepal_length'], bins=20, kde=True, color='blue')
plt.title('Sepal Length')

plt.subplot(2, 2, 2)
sns.histplot(df['sepal_width'], bins=20, kde=True, color='orange')
plt.title('Sepal Width')

plt.subplot(2, 2, 3)
sns.histplot(df['petal_length'], bins=20, kde=True, color='green')
plt.title('Petal Length')

plt.subplot(2, 2, 4)
sns.histplot(df['petal_width'], bins=20, kde=True, color='red')
plt.title('Petal Width')

plt.show()
```

```python
#산점도 시각화(이상치 판별 가능)
sns.scatterplot(x='sepal_length', y='petal_length', data=df, hue='species', palette='Set2')
plt.title('Scatter plot of Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend(title='Species')
plt.show()
```
