```python
#실습문제 : 다음은 한 학생의 일주일 간의 공부 시간과 성적 데이터
# 왼쪽 데이터를 이용하여 데이터 프레임을 생성하시오.
import pandas as pd
data = ([['월', 2, 70], ['화', 3, 75], ['수', 1, 65], ['목', 5, 85], ['금', 4, 80], ['토', 2, 70], ['일', 1,60]])
df = pd.DataFrame(data, columns = ['요일', '공부시간', '성적'])
df
```

```python
plt.rc('font', family='NanumGothic')
plt.plot(df['요일'], df['공부시간'], marker = 'o')
plt.show
```

```python
plt.rc('font', family='NanumGothic')
plt.plot(df['요일'], df['성적'], marker = 'o')
plt.show
```

```python
#생성한 데이터프레임을 이용하여 공부 시간과 성적 사이를 산점도로 나타내세요. x축은 공부시간, y축은 성적으로 설정하시오.
plt.rc('font', family='NanumGothic')
plt.scatter(df['공부시간'], df['성적'])
plt.show
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_excel('data1.xlsx')
plt.plot(df['name'], df['kor'], 'g--', marker = 'o', label = '국어점수')
plt.plot(df['name'], df['math'], 'r', marker = 'v', label = '수학점수')
plt.rc('font', family='NanumGothic')
plt.title('성적 그래프')
plt.xlabel('이름')
plt.ylabel('성적')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = pd.DataFrame({
'category': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'C', 'B', 'B'],
'value': [10, 20, 15, 25, 30, 12, 18, 22, 28, 35]
})
sns.countplot(x='category',hue = 'category', data = data)
plt.show()
```
